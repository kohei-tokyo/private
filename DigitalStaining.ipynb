```python
test_names = ["NoDiscriminator", "ResnetPatch", "Patch4"]
folders = ["1", "2", "3", "4", "5"]
n = 5

for test_name in test_names:
    for i in range(n):
        torch.cuda.memory.empty_cache()
        
        # 現在のイテレーション番号に基づいてフォルダを分割
        test_f = [folders[i]]
        val_f = [folders[(i + 1) % n]] # テストフォルダの次を検証用にする（リストの最後に達したら最初に戻る）
        train_f = ["c" + f for f in folders if f not in test_f and f not in val_f]
        
        print(f"--- Running for {test_name}, Fold {i+1} ---")
        print(f"Train: {train_f}, Val: {val_f}, Test: {test_f}")
        
        ds = DigitalStaining(
            dir=r"D:\Matsusaka\data_mito\COS7_KDEL-mSG_kmeans",
            name=f"{str(i)}_{test_name}_ER_251009",
            train_folders=train_f,
            val_folders=val_f,
            test_folders=test_f,
            n_epoch=100,
            discriminator=test_name
        )
        ds.all() # モデルの学習と評価を実行
torch.cuda.memory.empty_cache()

for test_name in test_names:
    for i in range(n):
        torch.cuda.memory.empty_cache()
        
        # 現在のイテレーション番号に基づいてフォルダを分割
        test_f = [folders[i]]
        val_f = [folders[(i + 1) % n]] # テストフォルダの次を検証用にする（リストの最後に達したら最初に戻る）
        train_f = ["c" + f for f in folders if f not in test_f and f not in val_f]
        
        print(f"--- Running for {test_name}, Fold {i+1} ---")
        print(f"Train: {train_f}, Val: {val_f}, Test: {test_f}")
        
        ds = DigitalStaining(
            dir=r"D:\Matsusaka\data_mito\HeLa_Su9-mSG_kmeans",
            name=f"{str(i)}_{test_name}_mito_251009",
            train_folders=train_f,
            val_folders=val_f,
            test_folders=test_f,
            n_epoch=100,
            discriminator=test_name
        )
        ds.all() # モデルの学習と評価を実行
torch.cuda.memory.empty_cache()
```
